```python
# importe a biblioteca do pandas
import pandas as pd
```

```python
#utilizando a url do site terra, transforme as informações em um dataframe
url_terra = 'https://www.cbf.com.br/futebol-brasileiro/competicoes/campeonato-brasileiro-serie-a/2022'

html_terra = pd.read_html(url_terra)
tabelaterra = html_terra[0]
display(tabelaterra)
```

```python
#utilizando a url do site da cnn, transforme as informações em um dataframe
url_cnn = 'https://www.cnnbrasil.com.br/esporte/futebol/tabela-brasileirao-serie-a-2022/'

html_cnn = pd.read_html(url_cnn)
tabelacnn = html_cnn[0]
display(tabelacnn)
```

```python
#junte os dois dataframe criados em um novo dataframe chamado dt_final
dt_final = pd.concat([tabelaterra,tabelacnn],axis=1)
display(dt_final)
```

```python
#Neste momento responda:
#Quantas linhas e colunas tem o dataframe criado pela junção? Coloque o comando utilizado para saber os valores aqui
print(dt_final.shape)
# 20 linhas e 23 colunas
```

```python
# Remova as colunas do dt_final com os seguintes nomes ['Classificação', 'PTS', 'j', 'V', 'e', 'd', 'g','GC', 'SG', 'Próx', '%', 'Recentes', 'CV']
dt_final2 = dt_final.drop(['Classificação', 'PTS', 'j', 'V', 'e', 'd', 'g','GC', 'SG', 'Próx', '%', 'Recentes', 'CV'],axis = 1)
display(dt_final2)
```
