```python
import pandas as pd
```

```python
url = 'https://www.cnnbrasil.com.br/esporte/futebol/tabela-brasileirao-serie-a-2022/'
html_cnn = pd.read_html(url)
tabelacnn = html_cnn[0]
display(tabelacnn)
```

```python
tabelacnn['UF'] = ['SP','RS','RJ','SP','RJ','PR','MG','CE','SP','MG','RJ','SP','GO','SP','PR','MT','CE','GO','SC','RS']
display(tabelacnn)
```

```python
tabelacnn['Títulos'] = ['10','3','4','7','7','1','2','0','6','0','2','8','0','0','1','0','0','0','0','0']
display(tabelacnn)
```

```python
tabelacnn.to_excel('tabelacnn.xlsx', index = False)
```
